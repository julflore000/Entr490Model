```python

#loading in respective imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
sns.set(style="darkgrid")
```

```python
dfPercent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT40GT80.xlsx",sheet_name="percentGeneration")

dfCT40GT80Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT40GT80.xlsx",sheet_name="percentGeneration")

dfCT40GT100Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT40GT100.xlsx",sheet_name="percentGeneration")

dfCT40GT115Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT40GT115.xlsx",sheet_name="percentGeneration")

dfCT100GT80Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT100GT80.xlsx",sheet_name="percentGeneration")

dfCT100GT100Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT100GT100.xlsx",sheet_name="percentGeneration")

dfCT100GT115Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT100GT115.xlsx",sheet_name="percentGeneration")

dfCT200GT80Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT200GT80.xlsx",sheet_name="percentGeneration")

dfCT200GT100Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT200GT100.xlsx",sheet_name="percentGeneration")

dfCT200GT115Percent = pd.read_excel("../../modelOutputs/outputEnergyCommitmentCT200GT115.xlsx",sheet_name="percentGeneration")

```

```python
dfCT200GT80Percent.head()
```

```python
#now looking at histplot
#combining all onto density plot
graph = sns.kdeplot(data=dfCT40GT80Percent*100, x="Solar",color="Black")
sns.kdeplot(data=dfCT100GT80Percent*100, x="Solar",color="Gray")
sns.kdeplot(data=dfCT200GT80Percent*100, x="Solar",color="Green")
graph.set(xlabel = "Percent of Optimal Energy Mix at hour",title="Solar: Kernel density estimates Percent of Optimal Energy Mix 2021")
plt.legend(labels=["CT40", "CT100", "CT200"])
```

```python
#combining all onto density plot for wind
graph = sns.kdeplot(data=dfCT40GT80Percent*100, x="Wind",color="Black")
sns.kdeplot(data=dfCT100GT80Percent*100, x="Wind",color="Gray")
sns.kdeplot(data=dfCT200GT80Percent*100, x="Wind",color="Green")
graph.set(xlabel = "Percent of Optimal Energy Mix at hour",title="Wind: Kernel density estimates Percent of Optimal Energy Mix 2021")
plt.legend(labels=["CT40", "CT100", "CT200"])
```

```python
#now looking at NG
#combining all onto density plot for wind
graph = sns.kdeplot(data=dfCT40GT80Percent*100, x="NG Gas",color="Black")
sns.kdeplot(data=dfCT100GT80Percent*100, x="NG Gas",color="Gray")
graph.set(xlabel = "Percent of Optimal Energy Mix at hour",title="NG: Kernel density estimates Percent of Optimal Energy Mix 2021")
plt.legend(labels=["CT40", "CT100"])
```

```python
#now looking at Geothermal
#combining all onto density plot for wind
graph = sns.kdeplot(data=dfCT200GT80Percent*100, x="Geothermal",color="Green")

graph.set(xlabel = "Percent of Optimal Energy Mix at hour",title="Geothermal: Kernel density estimates Percent of Optimal Energy Mix 2021")
plt.legend(labels=["CT200"])
```

```python
#Now looking at varying geothermal LCOEs at fixed carbon tax
graph = sns.kdeplot(data=dfCT200GT80Percent*100, x="Wind",color="Black")
sns.kdeplot(data=dfCT200GT100Percent*100, x="Wind",color="Gray")
sns.kdeplot(data=dfCT200GT115Percent*100, x="Wind",color="Green")
graph.set(xlabel = "Percent of Optimal Energy Mix at hour",title="Wind: Kernel density estimates Percent of Optimal Energy Mix 2021")
plt.legend(labels=["GT LCOE: 80", "GT LCOE: 100", "GT LCOE: 115"])
plt.yticks(graph.get_yticks(), graph.get_yticks())
plt.ylabel('Y', fontsize=7)
```
